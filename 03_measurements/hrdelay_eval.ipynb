```python
fname = "./hrdelay/hrdelay2.log"
bin_size = pow(2,8)

bins2 = np.empty(200)

with open(fname,'r') as f:
    for line in f:
        if line.startswith('#'):
            pass
        else:
            tags = line.split()
            result = parse('[bin{}]',tags[0]);
            bin_idx = int(result[0]);
            cnt = int(tags[1]);
            bins2[bin_idx] = cnt;
            #print "%d - %d" % (bin_idx, cnt)
            
bins_start2 = np.empty(200);
for i in range(200):
    bins_start2[i] = bin_size*i;
    
# calculate mean delay
mean = (sum(bins_start2 * bins2))/(sum(bins2))
 
fig = figure(figsize=[5,2])
plot( bins_start2/1000, bins2,'x')     
yscale('log')
title('High resolution timer callback delay in kernel space')
xlabel('Delay from requested time. [usec]')
ylabel('Count')
grid()
text(35,1e4,'Mean:%.2fusec' % (mean/1000), ha='left', bbox=dict(boxstyle="square", ec=(1., 0.5, 0.5),  fc="yellow"))
fig.savefig('./hrdelay/delay',bbox_inches = "tight")
```
