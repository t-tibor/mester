```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from matplotlib import gridspec
```

```python
fname = './xenomai_latency/max.txt';
stress_data = np.genfromtxt(fname,delimiter = ';', skip_header = 1);
max_latency = stress_data[:,0];
max_count = stress_data[:,1];

fname = './xenomai_latency/avg.txt';
stress_data = np.genfromtxt(fname,delimiter = ';', skip_header = 1);
avg_latency = stress_data[:,0];
avg_count = stress_data[:,1];

fname = './xenomai_latency/min.txt';
stress_data = np.genfromtxt(fname,delimiter = ';', skip_header = 1);
min_latency = stress_data[:,0];
min_count = stress_data[:,1];
```

```python
plot(max_latency,max_count,'x-')
xlabel('Delay [us]');
ylabel('Count');
title('Xenomai built-in latency test - max values');
```

```python
plot(avg_latency,avg_count,'x-')
xlabel('Delay [us]');
ylabel('Count');
title('Xenomai built-in latency test - avg values');
```

```python
plot(min_latency,min_count,'x')
xlabel('Delay [us]');
ylabel('Count');
title('Xenomai built-in latency test - min values');
```
