```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from matplotlib import gridspec
%matplotlib inline
```

```python
fname = "./motor_ident/motor.log";
T = np.genfromtxt(fname,dtype='uint64',delimiter=',',skip_header = 3, skip_footer=1);
ts_raw = T[:,0];
ts = (ts_raw - ts_raw[0])*1e-9;
T = np.genfromtxt(fname,dtype='float',delimiter=',',skip_header = 3, skip_footer=1);
period = T[:,1];
error = T[:,2];
prop = T[:,3];
integral = T[:,4];
duty = T[:,5];
```

```python
f,ax1 = subplots(figsize=(15,10));
ax1.plot(ts, 1/period,'x-',label='speed')
ax1.set_ylim([20,26])

ax2 = ax1.twinx()
ax2.plot(ts, duty,'-',Color = 'y', label='duty')
legend();
grid();
```

```python
f2,ax1 = subplots(figsize=(15,10));
ax1.plot(ts, 1/period,'x-',label='speed')
ax1.set_ylim([20,26])

ax2 = ax1.twinx();
ax2.plot(ts, integral,'-',Color = 'y', label='integral')
legend();
grid();
```
